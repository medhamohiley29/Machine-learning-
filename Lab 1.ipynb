```python
import numpy as np

myarray13 = np.array([13, 24, 22, 13, 11, 28, 16, 24, 18], int)

print("Original array is:", myarray13)

index = 3
new_value = 45
myarray13[index] = new_value

print("Modified array is:", myarray13)
print("Size of the array is:", myarray13.size)
print("Data typ of the array is:", myarray13.dtype)
print("Mean of the array is:", np.mean(myarray13))
print("Median of the array is:", np.median(myarray13))
print("Minimum of all array elements is:", np.min(myarray13))
print("Maximum of all array elements is:", np.max(myarray13))
print("Sum of all array elements is:", np.sum(myarray13))
print("Product of all array elements is:", np.prod(myarray13))
print("Co-variance of all array elements is:", np.cov(myarray13))
print("Variance of all array elements is:", np.var(myarray13))
print("Standard deviation of array elements is:",np.std(myarray13))
print("The sorted array is:", np.sort(myarray13))
print("Power raised to element(3) of array:", np.power(myarray13,3))
print("Square root of array elements is:",np.sqrt(myarray13))
print("Absolute value of array elements is:", np.abs(myarray13))
print("Sin value of array elements is:", np.sin(myarray13))
print("Cosine of array elements is:", np.cos(myarray13))
print("Tangent of array elements is:", np.tan(myarray13))
print("Log of all elements with e base is", np.log(myarray13))
print("Log of all elements with base 10 is:", np.log10(myarray13))
print("Exponintial of the array elements is:", np.exp(myarray13))
```
